### Import

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Read the data to X and y, and replace B and M (2&4) with 0 and 1

```python
data = pd.read_csv("data.csv")
X = data.drop("Class", axis=1)
y = data['Class']
y.replace({2: 0, 4: 1}, inplace=True)

data.info()
```

** Note that there are missing values in Bare_nuclei **

### Box plot

```python
# set theme
sns.set_theme(style="whitegrid")

fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(10, 8))
ax = ax.flatten()

for i, column in enumerate(X.columns):
    sns.boxplot(x=y, y=X[column], ax=ax[i])
    ax[i].set_title(f'{column} vs Class')
    ax[i].set_xlabel('Class')
    ax[i].set_ylabel(column)

plt.tight_layout()
plt.show()
```

### Kernel Density Estimate (KDE) Plot

```python
# set theme
sns.set_theme(style="whitegrid")

fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(10, 8))
ax = ax.flatten()

for i, column in enumerate(X.columns):
    sns.kdeplot(x=y, y=X[column], ax=ax[i], alpha=0.5)
    ax[i].set_title(f'{column} vs Class')
    ax[i].set_xlabel('Class')
    ax[i].set_ylabel(column)
    ax[i].set_xlim(-1, 2)

plt.tight_layout()
plt.show()
```
